```python
# Import the pathlib and csv library
from pathlib import Path
import csv

# Check the current directory where the Python program is executing from
print(f"Current Working Directory: {Path.cwd()}")

# Set the path using Pathlib
file = Path('../budget_data.csv')

# Initiate Variables
month_count = 0
pnl = 0

#Initiate List to hold monthly profit/loss
profitloss = []

#Initiate dictionary to hold summary
fin_summary = {}

# Open the csv file as an object
with open(file, 'r') as csvfile:
    
# prints the data type of the file--->    print(type(csvfile))

# Pass in the csv file to the csv.reader() function
# (with ',' as the delmiter/separator) and return the csvreader object
    csvreader = csv.reader(csvfile, delimiter=",")
    print(type(csvreader))

    csvheader = next(csvreader)
#Skipping the print of the header--->    print(csvheader)

 # Read each row of data after the header
    for row in csvreader:
        # Print the row
        #prints the contents of each row--->  print(row) 
 
    # Set the date count variable, month_count and add monthly values to the list & calculate total profit/loss
       
          date = row[0]
       
          month_count += 1
        
          profitloss.append(row)
        
          pnl += int(row[1])
        
   # To check the mid way status of total months---> print(month_count)
   # To check the mid way status of total revenue---> print(pnl)
    
  # Initialize variables to hold monthly average, monthly total, month-on-month change, max increase and max decrease values 
    monthly_avg = 0
    monthly_total = 0
    monm = 0
    max_dec = 0
    max_inc = 0
   
  
        
        # Compute the change in "Profit/Losses" between months over the entire period
        # by iterating from the start
    for x in range(0, month_count -1, 1):
        monm = int(profitloss[x+1][1])- int(profitloss[x][1])
      
    
    # Find the values of Greatest Increase (max_increase) and Greatest Decrease (max_decrease)
        if monm < max_dec:
            min_month = profitloss[x + 1][0]
            max_dec = monm
        elif monm > max_inc:
            max_month=profitloss[x + 1][0]
            max_inc = monm
            
     # Total change in Profit/Loss between months over the entire period
        monthly_total += monm
    print(monthly_total)
       
    

#The average change in monthly Profit/Loss over the entire period    
    
    monthly_avg = round(monthly_total/(month_count-1),2)
    
    #Print the Financial Analysis to the terminal
print("Financial Analysis")
print("----------------------------")
print(f"Total Months:  {month_count}")
print(f"Total:  ${pnl}")
print(f"Average Change:  ${monthly_avg}")
print(f"Greatest Increase in Profits:  {max_month} (${max_inc})")
print(f"Greatest Decrease in Profits:  {min_month} (${max_dec})")
            
        
# Exports the results to the text file
output_path = Path("output_budget_data.txt")

# Open the output path as a file object
with open(output_path, 'w') as file:
    file.write("Financial Analysis\n")
    file.write("----------------------------\n")
    file.write(f"Total Months:  {month_count}\n")
    file.write(f"Total:  ${pnl}\n")
    file.write(f"Average Change:  ${monthly_avg}\n")
    file.write(f"Greatest Increase in Profits:  {max_month} (${max_inc})\n")
    file.write(f"Greatest Decrease in Profits:  {min_month} (${max_dec})\n")

        
```
